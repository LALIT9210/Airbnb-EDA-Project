## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

**Start here Vizualization the data..**

```python
for column in numerical_column[2:]: # this chart shows the mean and median of all values
   fig = plt.figure(figsize=(9, 6))
   ax = fig.gca()
   feature = df[column]
   feature.hist(bins=100, ax = ax)
   ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
   ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
   ax.set_title(column)
plt.show()
```

From the distribution of filtered numerical_column data. Median as well Mean tell use the avg value of all the column and median tell 50% of the value so the general all the values lie between them. due to this we can findout the general idea of all the values like avg of pricing, minimum nights, reviews etc., it means we have all sort of listing available uniformly throughout the year.

```python
# People are private rooms preferred over other room types?
df_new['room_type'].value_counts() # so according to this people preferred  more Entire home/apt room compare to others.
```

```python
# How many hotel available 365 days? And top 10 hotels name on the basis of highest availability for the whole year?
df1= df_new[df_new['availability_365']==365]
df1['name'].value_counts().sum() # total number of host available for 365 days
```

```python
# This tells about available host name who open for 365 days with the room types list of 10 hotels
hosts_available_365days = df_new.groupby(['name','room_type','host_id'])['availability_365'].max().reset_index()
hosts_available_365days.sort_values(by='availability_365',ascending=False).head(10)
```

```python
id = hosts_available_365days['host_id'].head(5)
availability = hosts_available_365days['availability_365'].head(5)
names = hosts_available_365days['name'].head(5)

fig, ax = plt.subplots(figsize = (10, 5))

# creating the scatter plot
ax.scatter(availability, id)

# adding labels to data points
for i, txt in enumerate(names):
    ax.annotate(txt, (availability[i], id[i]))

plt.xlabel("Available 365 days")
plt.ylabel("Host ID")
plt.title("Availability vs Host ID")
plt.show()
```

In there we creates a scatter plot with annotations to visualize the relationship between the availability of hosts for 365 days and their corresponding host IDs. This plot aims to provide insights into the availability patterns of the first five hosts in the dataset.


```python
# Correlation between all the numerical data values
plt.figure(figsize=(15,8))
correlation = df_new.corr()
sns.heatmap(abs(correlation), annot=True, cmap='coolwarm')
```

no_of_reviews and reviews_per_month has high co-relation for obvious reason. But the price column also has very low corelation with other features. Let check it how it differes or related.

```python
# Minimum nights people should pay per hotels
nights = df_new.groupby(['minimum_nights','name'])['price'].max().reset_index()
nights.sort_values(by='price',ascending=False).head()
```

```python
# most busiest hosts and why?
busiest_host = df_new.groupby(['host_name','host_id'])['number_of_reviews'].max().reset_index()
busiest_host.sort_values(by='number_of_reviews',ascending=False).head(10)
```

**Dona,Jj,Maya,Carol,Danielle is the top3 busiest hotels.**

```python
#Create a bar plot
# Select the data to plot
name = df_new['host_name'].head(10)
reviews = df_new['number_of_reviews'].head(10)

# Create a bar chart using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(12,7))
sns.barplot(x=name,y=reviews,color = 'gold',width=0.3)
plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts")

plt.show()
```

This graph basically shows the most active host on Airbnb platform because its number of reviews is highest.

```python
# Which area get highest number of reviews
highest_reviews_area = df_new.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
highest_reviews_area.sort_values(by='number_of_reviews',ascending=False)
```

```python
area =highest_reviews_area['neighbourhood_group']
review =highest_reviews_area['number_of_reviews']
fig=plt.figure(figsize=(10,5))

# creating the bar plot
plt.bar(area, review, color ='green', width=0.3)

plt.xlabel("area")
plt.ylabel("review")
plt.title("Area vs Number of reviews")
plt.show()
```

This graph illustrates the distribution of reviews across different areas, allowing us to identify the most popular location or the one with the most favored amenities compared to others.

```python
price_area = df_new.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head()
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize=(10,5))

# creating the scatter plot
plt.scatter(area, price)
plt.xlabel("Price")
plt.ylabel("number_of_reviews")
plt.title("Price vs Number of Reviews")
plt.show()
```

This graph depicts the correlation between the number of reviews and pricing categories. It indicates whether guests prefer expensive or budget-friendly hosts. From the graph, it is evident that a significant portion of the hosts falls into the moderate pricing range, suggesting that this category is preferred by most guests.

```python
# provides an overview of how properties are distributed across various neighborhoods.
distribution = df_new.groupby(['calculated_host_listings_count'])['neighbourhood'].max().reset_index()
distribution.sort_values(by='neighbourhood',ascending=True).head()
```

```python
# set plot background grid style
sns.set(style = "darkgrid")

# set the figure size
plt.figure(figsize=(10,7))

# plot a bar chart
sns.barplot(x="calculated_host_listings_count", y="neighbourhood",data=distribution,estimator=sum,ci=None,color='gold')
```

Based on this distribution, it is evident that the majority of clients have listed their properties in the Upper East Side neighborhood, while the fewest listings are found in Windsor Terrace.

```python
# Room type availability according to their percentage?
df_new["room_type"].value_counts().plot(kind = "pie",autopct = "%0.1f%%" )
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```


This donut chart shows the proportion of each room type's availability relative to the total, making it easy to understand the percentage distribution of room types

```python
# Grouping neighbourhood according to location with pricing
grouped = df_new.groupby('neighbourhood_group')

for group_name, group_data in grouped:
    top_3 = group_data.nlargest(3, 'price')
    print(f'Top 3 neighborhoods in {group_name}:')
    print(top_3[['neighbourhood', 'price']])
```

```python
fig = plt.figure(figsize = (10, 5))
for group_name, group_data in grouped:
    top_3 = group_data.nlargest(3, 'price')

    plt.figure()
    plt.pie(top_3['price'], labels=top_3['neighbourhood'], autopct='%1.1f%%' )
    plt.title(f'Top 3 neighborhoods in {group_name}')
    plt.show()
```

```python
# Most demanding Host for Airbnb
host_based_on_review_df = df_new.groupby(['host_id','host_name'],as_index=False)['number_of_reviews'].sum()
host_based_on_review_df.sort_values(['number_of_reviews'],ascending = False)
host_based_on_availability_df = df_new.groupby(['host_id','host_name'],as_index=False)['availability_365'].mean()
host_based_on_availability_df.sort_values(['availability_365'],ascending = True)
df_new1 = pd.merge(host_based_on_review_df,host_based_on_availability_df,on = ['host_id','host_name'])
df_new1.sort_values(['availability_365','number_of_reviews'],ascending = [True,False]).head(5)
```

```python
df_new1.plot.scatter(x='number_of_reviews', y='availability_365', xlim=(100,1000),ylim=(10,380),color='gold');
```

As we cann see from this scatter plot that highest number of reviews given by client is to those host who available maximum days.
